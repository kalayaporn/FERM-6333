```python
# Removing non-numerical variables

dataset = data.select_dtypes(include = [np.number])
```

```python
# Removing variables (Column) with more than 70% missing

wdrs = dataset[dataset.columns[dataset.isnull().sum()/dataset.shape[0]<0.7]]
wdrs
```

```python
# Filling the missing value with median of each column

wdrsdata = wdrs.fillna(wdrs.median())
wdrsdata
```

```python
# Checking missing value for each variables (column)

wdrsdata.isnull().sum()
```

```python
# X is a predictor variable or an independent variable
# Y is a dependent variable or an outcome variable

X = wdrsdata[wdrsdata.columns[~wdrsdata.columns.isin(['ebit'])]]
y = wdrsdata[wdrsdata.columns[wdrsdata.columns.isin(['ebit'])]]
```

```python
'''
Perform a forward-backward feature selection 
    based on p-value from statsmodels.api.OLS
    Arguments:
        X - varaiables without predictor variable (EBIT)
        y - Predictor Varaible ; EBIT
        Initial - list of features to start with (column names of X)
        ListIn - include a feature if its p-value < ListIn
        ListOut - exclude a feature if its p-value > ListOut
        A_D - whether to print the sequence of inclusions and exclusions
    Returns: list of selected features 
    Always set threshold_in < threshold_out to avoid infinite looping.
'''

def stepwise_selection(X, y, 
                       Initial=[], 
                       ListIn =0.01, 
                       ListOut = 0.05, 
                       A_D=True):
    included = list(Initial)
    while True:
        changed=False
        # Using forward step
        excluded = list(set(X.columns)-set(included))
        new_pval = pd.Series(index=excluded)
        for new_column in excluded:
            model = sm.OLS(y, sm.add_constant(pd.DataFrame(X[included+[new_column]]))).fit()
            new_pval[new_column] = model.pvalues[new_column]
        best_pval = new_pval.min()
        if best_pval < ListIn:
            best_feature = new_pval.argmin()
            included.append(best_feature)
            changed=True
            if A_D:
                print('Add  {:30} with p-value {:.6}'.format(best_feature, best_pval))
        # Using "backward step"
        model = sm.OLS(y, sm.add_constant(pd.DataFrame(X[included]))).fit()
        # use all coefs except intercept
        pvalues = model.pvalues.iloc[1:]
        worst_pval = pvalues.max() # null if pvalues is empty
        if worst_pval > ListOut:
            changed=True
            worst_feature = pvalues.argmax()
            included.remove(worst_feature)
            if A_D:
                print('Drop {:30} with p-value {:.6}'.format(worst_feature, worst_pval))
        if not changed:
            break
    return included

result = stepwise_selection(X, y)
# Rejecting the H0 : when p-val is greater than 0.05 or 5%
```

```python
# Printing the results

print('resulting features:')
print(result)
```
