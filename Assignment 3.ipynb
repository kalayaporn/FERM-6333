```python
# Cross-validation:
# Split a dataset into k folds

def cross_validation_split(dfw, n_folds):
    wdrsdata_split = list()
    wdrsdata_copy = list(wdrsdata)
    fold_size = int(len(wdrsdata) / n_folds)
    for i in range(n_folds):
        fold = list()
        while len(fold) < fold_size:
            index = randrange(len(wdrsdata_copy))
            fold.append(wdrsdata_copy.pop(index))
        wdrsdata_split.append(fold)
    return wdrsdata_split

# Calculating accuracy percentage

def accuracy_metric(actual, predict):
    correct = 0
    for i in range(len(actual)):
        if actual[i] == predict[i]:
            correct += 1
    return correct / float(len(actual)) * 100.0

# Using a cross validation split for evaluating an algorithm

def evaluate_algorithm(wdrsdata, algorithm, n_folds, *args):
    folds = cross_validation_split(wdrsdata, n_folds)
    point = list()
    for fold in folds:
        train_set = list(folds)
        train_set.remove(fold)
        train_set = sum(train_set, [])
        test_set = list()
        for row in fold:
            row_copy = list(row)
            test_set.append(row_copy)
            row_copy[-1] = None
        predict = algorithm(train_set, test_set, *args)
        actual = [row[-1] for row in fold]
        accuracy = accuracy_metric(actual, predict)
        point.append(accuracy)
    return point
```

```python
# Creating split:
# Spliting the dataset based on an attribute and an attribute value

def test_split(index, value, wdrsdata):
    left, right = list(), list()
    for row in dataset:
        if row[index] < value:
            left.append(row)
        else:
            right.append(row)
    return left, right

# Selecting the best split point for the dataset

def get_split(wdrsdata):
    class_values = list(set(row[-1] for row in wdrsdata))
    b_index, b_value, b_point, b_groups = 999, 999, 999, None
    for index in range(len(wdrsdata[0])-1):
        for row in wdrsdata:
            groups = test_split(index, row[index], wdrsdata)
            gini = gini_index(groups, class_values)
            if gini < b_point:
                b_index, b_value, b_point, b_groups = index, row[index], gini, groups
    return {'index':b_index, 'value':b_value, 'groups':b_groups}
 
# Creating a terminal node value

def to_terminal(group):
    outcomes = [row[-1] for row in group]
    return max(set(outcomes), key = outcomes.count)
 
# Creating child splits for a node or make terminal
# Checking both left and right Child

def split(node, max_depth, min_size, depth):
    left, right = node['groups']
    del(node['groups'])
    # check for a no split
    if not left or not right:
        node['left'] = node['right'] = to_terminal(left + right)
        return
    # Checking for max depth
    if depth >= max_depth:
        node['left'], node['right'] = to_terminal(left), to_terminal(right)
        return
    # Processing left child
    if len(left) <= min_size:
        node['left'] = to_terminal(left)
    else:
        node['left'] = get_split(left)
        split(node['left'], max_depth, min_size, depth + 1)
    # Processing right child
    if len(right) <= min_size:
        node['right'] = to_terminal(right)
    else:
        node['right'] = get_split(right)
        split(node['right'], max_depth, min_size, depth + 1)
```

```python
# Creating Gini Index:
# Calculating the Gini index for the split dataset

def gini_index(groups, classes):
    # count all samples at split point
    n_instances = float(sum([len(group) for group in groups]))
    # sum weighted Gini index for each group
    gini = 0.0
    for group in groups:
        size = float(len(group))
        # avoid divide by zero
        if size == 0:
            continue
        score = 0.0
        # score the group based on the score for each class
        for class_val in classes:
            p = [row[-1] for row in group].count(class_val) / size
            score += p * p
        # weight the group score by its relative size
        gini += (1.0 - score) * (size / n_instances)
    return gini
```

```python
# Decision tree:
# Building a decision tree

def build_tree(train, max_depth, min_size):
    root = get_split(train)
    split(root, max_depth, min_size, 1)
    return root
 
# Making a prediction with a decision tree

def predicted(node, row):
    if row[node['index']] < node['value']:
        if isinstance(node['left'], dict):
            return predicted(node['left'], row)
        else:
            return node['left']
    else:
        if isinstance(node['right'], dict):
            return predicted(node['right'], row)
        else:
            return node['right']

# Classification and Regression Tree Algorithm 

def decision_tree(train, test, max_depth, min_size):
    tree = build_tree(train, max_depth, min_size)
    predictions = list()
    for row in test:
        prediction = predicted(tree, row)
        predictions.append(prediction)
    return(predictions)
```
