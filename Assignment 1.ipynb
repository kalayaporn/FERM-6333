```python
dataset= data[data.columns[data.isnull().sum()/data.shape[0]<0.5]]
```

```python
# Showing the top part of dataset
dataset.head()
```

```python
dataset.describe(include = 'all').style
```

```python
def color_type(val):
    color = 'blue' if val < 0 else 'black'
    return 'color: %s' % color
```

```python
table=dataset.describe().style.applymap(color_type)
table
```

```python
# Numeric variables  table 
cm = sns.light_palette("pink", as_cmap=True)
table.background_gradient(cmap=cm)
```

```python
# Categorical variables statistics table
dataset.describe(include=['O']).style.set_properties(**{'background-color': 'grey',
                           'color': 'black',
                           'border-color': 'white'})
```
